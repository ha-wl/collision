```python
import pandas as pd
import numpy as np
from pathlib import Path

date = ("2024-12-02/20-13-14")
path1 = Path("../gen_outputs/sonoisa__t5-base-japanese") / date / "test-metrics.jsonl"
path2 = Path("../gen_outputs/sonoisa__t5-base-japanese") / date / "plain-metrics.jsonl"

df1 = pd.read_json(path1, orient='records', lines=True)
df2 = pd.read_json(path2, orient='records', lines=True)
#df1 = pd.read_json("../gen_outputs/sonoisa__t5-base-japanese/2024-06-01/10-56-59/test-metrics.jsonl", orient='records', lines=True)
#df2 = pd.read_json("../gen_outputs/sonoisa__t5-base-japanese/2024-06-01/10-56-59/plain-metrics.jsonl", orient='records', lines=True)
sum1 = 0
sum2 = 0
var_ave1 = 0
var_ave2 = 0
lst_length1 = 0
lst_length2 = 0

# 正規分布のKL divergence
def gaussian1d_KLdivergence(a1,b1,a2,b2):
    A = np.log(np.sqrt(b2)/np.sqrt(b1))
    B = ( b1 + (a1 - a2)**2 ) / (2*b2)
    C = -1/2
    y = A + B + C
    return y

for row1 in df1.itertuples():
    preds1 = row1.dist_pred
    lst1 = row1.dist_lst
    #print(preds[0])
    #print(preds[1])
    #print(lst[0])
    #print(lst[1])
    #print(gaussian1d_KLdivergence(preds[0],preds[1],lst[0],lst[1]))
    sum1 += gaussian1d_KLdivergence(preds1[0],preds1[1],lst1[0],lst1[1])
    var_ave1 +=preds1[1]
    lst_length1 += 1
    
for row2 in df2.itertuples():
    preds2 = row2.dist_pred
    lst2 = row2.dist_lst
    #print(preds[0])
    #print(preds[1])
    #print(lst[0])
    #print(lst[1])
    #print(gaussian1d_KLdivergence(preds[0],preds[1],lst[0],lst[1]))
    sum2 += gaussian1d_KLdivergence(preds2[0],preds2[1],lst2[0],lst2[1])
    var_ave2 += preds2[1]
    lst_length2 += 1

print("KL情報量の和と分散の平均")
print(sum1)
print(var_ave1/lst_length1)
print(sum2)
print(var_ave2/lst_length2)
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

#date = ("2024-06-09/09-50-29")
path = Path("../data/train7.jsonl")

df = pd.read_json(path, orient='records', lines=True)
#df1 = pd.read_json("../gen_outputs/sonoisa__t5-base-japanese/2024-06-01/10-56-59/test-metrics.jsonl", orient='records', lines=True)
#df2 = pd.read_json("../gen_outputs/sonoisa__t5-base-japanese/2024-06-01/10-56-59/plain-metrics.jsonl", orient='records', lines=True)
sum_we = 0
sum_sp = 0
lst_length1 = 0
lst_length2 = 0

# 正規分布のKL divergence
def gaussian1d_KLdivergence(a1,b1,a2,b2):
    A = np.log(np.sqrt(b2)/np.sqrt(b1))
    B = ( b1 + (a1 - a2)**2 ) / (2*b2)
    C = -1/2
    y = A + B + C
    return y

for row in df.itertuples():
    preds = row.weightspeed
    answer = row.weightspeed_answer
    sum_we += gaussian1d_KLdivergence(preds[0],preds[1],answer[0],answer[1]) + gaussian1d_KLdivergence(preds[2],preds[3],answer[2],answer[3])
    sum_sp += gaussian1d_KLdivergence(preds[4],preds[5],answer[4],answer[5]) + gaussian1d_KLdivergence(preds[6],preds[7],answer[6],answer[7])



print("KL情報量の和")
print(sum_we)
print(sum_sp)
```
